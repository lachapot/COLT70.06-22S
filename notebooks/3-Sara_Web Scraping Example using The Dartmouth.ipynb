```python
#alternatively, puts results into pandas to add metadata + context

# Step 1: Extract Metadata w/ Beautiful Soup

# gets the html for soup to look at
from bs4 import BeautifulSoup
import requests

URL = 'https://www.thedartmouth.com/article/2022/04/inflation-labor-shortages-taking-toll-on-hanover-businesses'
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")


# get headline
headline = soup.find('h1').text.strip()
print("Headline:", headline)

# get author
result = soup.find_all('a', href = True)
writer = ''

for r in result:
    if 'staff' in r['href']:
        writer = r.text
        print('Writer:', writer)
        break
        
# get date + time
time = soup.find('span', class_ = 'date-and-time').text.strip()
print("Time:", time)

# get tags
tags = soup.find('div', class_ = 'article-tags')
tag_list = [t.strip().replace(',', '') for t in tags.text.split( )][1:]
print('Tags:', tag_list)

# get article text
article = ''

article_bits = soup.find_all('p', class_ = None)

for a in article_bits:
    article += a.text.strip()
print('Article:', article)
```

```python
# Step 2: Store Metadata in Pandas!

import pandas as pd

data = {'Headline': headline,
        'Author':writer,
       'Date Published':time,
       'Tags':[tag_list],
       'Article':article}

df = pd.DataFrame(data)
df
```

```python
# Step 3: Repeat!

# gets all the links of articles 

from bs4 import BeautifulSoup
import requests

URL = 'https://www.thedartmouth.com/'
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")

result = soup.find_all('a', href = True)

URLs_to_visit = set()

for r in result:
    if 'article' in r['href']:
        URLs_to_visit.add(r['href'])
print(URLs_to_visit)


headlines = []
authors = []
dates = []
tagss = []
articles = []

for link in URLs_to_visit:
    URL = link
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")


    # get headline
    headline = soup.find('h1').text.strip()
    print("Headline:", headline)

    # get author
    result = soup.find_all('a', href = True)
    writer = ''

    for r in result:
        if 'staff' in r['href']:
            writer = r.text
            break

    # get date + time
    time = soup.find('span', class_ = 'date-and-time').text.strip()

    # get tags
    tags = soup.find('div', class_ = 'article-tags')
    tag_list = [t.strip().replace(',', '') for t in tags.text.split( )][1:]

    # get article text
    article = ''

    article_bits = soup.find_all('p', class_ = None)

    for a in article_bits:
        article += a.text.strip()
        
    headlines.append(headline)
    authors.append(writer)
    dates.append(time)
    tagss.append(tag_list)
    articles.append(article)

data = {'Headline': headlines,
        'Author':authors,
        'Tags':tagss,
       'Date Published':dates,
       'Article':articles}

df = pd.DataFrame(data)
df
```
